```python
# Importing the required libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the dataset

dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
# Getting the input data

x = dataset.iloc[:, [3,4]].values
x
```

```python
# Standardizing the input data

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_scaled = sc.fit_transform(x)
x_scaled
```

```python
# Creating the Affinity Propogation Model

from sklearn.cluster import AffinityPropagation
cluster_affinity = AffinityPropagation(affinity = 'euclidean')
y_pred = cluster_affinity.fit_predict(x_scaled)
y_pred
```

```python
dir(cluster_affinity)
```

```python
# Getting the cluster centers

cluster_centers = cluster_affinity.cluster_centers_
print("The Cluster centers are :", cluster_centers)
```

```python
# Printing the number of clusters

labels = cluster_affinity.labels_
n_clusters = len(np.unique(labels))
print("The Number of clusters formed :", n_clusters)
```

```python
labels
```

```python
cluster_centers_indices = cluster_affinity.cluster_centers_indices_
cluster_centers_indices
```

```python
# Evaluating the clustering using Silhouette Score

from sklearn.metrics import silhouette_score
s_score = silhouette_score(x_scaled, y_pred)
print("The Silhouette_Score is :", s_score)
```

```python
# Visualizing the clustering

fig = plt.figure()
#ax = fig.add_subplot(111)

colors = plt.cycler("color", plt.cm.viridis(np.linspace(0, 1, n_clusters)))

for cluster_num, col in zip(range (n_clusters), colors):
    cluster_points = labels == cluster_num
    cluster_centers = x_scaled[cluster_centers_indices[cluster_num]]
    plt.scatter(x_scaled[cluster_points, 0], x_scaled[cluster_points, 1], marker = '.', color = col["color"])
    plt.scatter(cluster_centers[0], cluster_centers[1], s = 55, marker = 'o',  color = col["color"])
    for x in x_scaled[cluster_points]:
        plt.plot([cluster_centers[0], x[0]], [cluster_centers[1], x[1]], color = col["color"])
       
    
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Affinity Propagation Clustering')
plt.show()    
```
